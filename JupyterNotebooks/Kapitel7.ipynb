```python
def Stellenwerte(n:int,b:int)->list[int]: 
    # Entwickelt n in Basis b
    r=[]
    while n>0:
        r=r+[n%b]
        n=n//b
    return r
print(Stellenwerte(7109,10))
print(Stellenwerte(7109,2))
def ZahlAusStellen(s:list[int],b:int)->int:
    return sum([s[k]*b**k for k in range(len(s))])
print(ZahlAusStellen([9,0,1,7],10))
```

```python
def plus(a:list,b:list,basis=10)->list: # Addition
    while len(a)<len(b):
        a=a+[0]
    while len(a)>len(b):
        b=b+[0]
    res=[]; übertrag=0
    for i in range(len(a)):
        s=a[i]+b[i]+übertrag
        res=res+[s%basis]
        übertrag=s//basis
    if übertrag>0:
        res=res+[übertrag]
    return res
a=Stellenwerte(7378,10); b=Stellenwerte(9876,10)
print(ZahlAusStellen(plus(a,b),10),"=",7378+9876)

def multZiffer(a:list,z:int,basis=10)->list: # Multipliziert mit einer Ziffer
    übertrag=0
    res=[]
    for k in range(len(a)):
        s=z*a[k]+übertrag
        res.append(s%basis)
        übertrag=s//basis
    if übertrag>0:
        res.append(übertrag)
    return res
print("321*4=",321*4,multZiffer([1,2,3],4))
def mult(a:list,b:list,basis=10)->list:
    res=[]
    for i in range(len(b)):
        res=plus(res,[0]*i+multZiffer(a,b[i],basis))
    return res
print(ZahlAusStellen(mult(a,b),10),"=",7378*9876)
print(mult(Stellenwerte(5,2),Stellenwerte(12,2),2))    
```

```python
# Brüche als Listen wie [1,2] für 1/2
def ggT(a:int,b:int)->int: # Variante mit Mehrfachzuweisungen
    while b !=0:
        [q, r] = [a//b, a%b] 
        [a,b]=[b,r]
    return a
def gekürzt(B:list[int])->list[int]:
    z=B[0]; n=B[1]
    if n<0:
        z=-z; n=-n
    g=ggT(abs(z),n)
    z=z//g; n=n//g
    return [z,n]
a=[5,6]; b=[12,-18]; print(b,"=",gekürzt(b))
def Badd(a:list[int],b:list[int])->list[int]:
    return gekürzt([a[0]*b[1]+a[1]*b[0],a[1]*b[1]])
print(a,"+",b,"=",Badd(a,b))
def Bsub(a:list[int],b:list[int])->list[int]:
    return gekürzt([a[0]*b[1]-a[1]*b[0],a[1]*b[1]])
print(a,"-",b,"=",Bsub(a,b))
def Bmul(a:list[int],b:list[int])->list[int]:
    return gekürzt([a[0]*b[0],a[1]*b[1]])
print(a,"*",b,"=",Bmul(a,b))
def Bdiv(a:list[int],b:list[int])->list[int]:
    return gekürzt([a[0]*b[1],a[1]*b[0]])
print(a,"/",b,"=",Bdiv(a,b))
```

```python
# Komplexe Zahlen als Listen wie [1,2] für 1+2i
import math
a=[5,6]; b=[12,-18];
def CBetrag(z:list[float])->float:
    return math.sqrt(z[0]**2+z[1]**2)
def CKonjug(z:list[float])->list[float]:
    return [z[0],-z[1]]
def Cadd(a:list[float],b:list[float])->list[float]:
    return [a[0]+b[0],a[1]+b[1]]
def Csub(a:list[float],b:list[float])->list[float]:
    return [a[0]-b[0],a[1]-b[1]]
print(a,"+",b,"=",Cadd(a,b))
print(a,"-",b,"=",Csub(a,b))
def Cmul(a:list[float],b:list[float])->list[float]:
    return [a[0]*b[0]-a[1]*b[1],a[0]*b[1]+a[1]*b[0]]
print(a,"*",b,"=",Cmul(a,b))
def Cdiv(a:list[float],b:list[float])->list[float]:
    return Cmul(a,Cinv(b))
def Cinv(a:list[float])->list[float]:
    n=a[0]**2+a[1]**2
    return [a[0]/n,-a[1]/n]
print(a,"/",b,"=",Cdiv(a,b))
```

```python
def Pstr(p:list,var="x")->str:
    if p==[]: return "0"
    s=""
    for i in range(len(p)-1,-1,-1):
        if p[i]==0:
            continue
        if i==0:
            s=s+"+"+str(p[i])
        if i==1:
            s=s+"+"+str(p[i])+"*"+var
        if i>1:
            if i==len(p)-1:
                s=str(p[i])+"*"+var+"^"+str(i)
            else: 
                s=s+"+"+str(p[i])+"*"+var+"^"+str(i)
    return s
print("Wert von ",Pstr(p),"bei x=2",Pwert(p,2))
```

```python
def Pstandard(p:list)->list:
    while len(p)>0 and p[-1]==0: 
        p=p[0:len(p)-1]
    return(p)
def Pdeg(p:list)->int: # Grad
    return len(Pstandard(p))-1
def Padd(a:list,b:list)->list: # Addition
    while len(a)<len(b):
        a=a+[0] 
    while len(b)<len(a):
        b=b+[0]         
    return Pstandard([a[i]+b[i] for i in range(len(a))])
print(Pstr(p)+" + "+Pstr(q)+" = "+Pstr(Padd(p,q)))
def Psub(a:list,b:list)->list: 
    while len(a)<len(b): # Subtraktion
        a=a+[0] 
    while len(b)<len(a):
        b=b+[0]         
    return Pstandard([a[i]-b[i] for i in range(len(a))])
def PSmul(p:list, s)->list: # Multiplikation mit Zahl s
    return Pstandard([s*x for x in p]) 
def Pmul(a:list,b:list)->list:
    res=[0]*(1+(len(a)-1)+(len(b)-1)) 
    # Ergebnis wird Polynom vom Grad(a)+Grad(b)
    for i in range(len(a)):
        for j in range(len(b)):
            res[i+j]+=a[i]*b[j]
    return Pstandard(res)
print(Pstr(p)+" * "+Pstr(q)+" = "+Pstr(Pmul(p,q)))
```

```python
def Pdiv(dividend:list, divisor:list)->list[list]:
    def quot(a,b): # Division, ggf als Bruch
        if type(a)==int and type(b)==int: 
            if a%b==0: 
                return a//b
            else:
                return Fraction(a,b)
        return a/b
    # Kopien anlegen und führende Nullen entfernen
    r=Pstandard(dividend[:]); d=Pstandard(divisor[:])
    if len(d)==0 or (len(d) == 1 and d[0] == 0):
        raise ZeroDivisionError("Division durch das Nullpolynom")
    m = len(r)-1 # Grad dividend
    n = len(d)-1 # Grad divisor
    if m<n: return [[0],r]
    q=[0]*(m-n+1) # Liste passender länge für den Quotienten
    while len(r)-1>=n:
        coef=quot(r[-1],d[-1])
        deg =(len(r)-1)-n # Grad des Ergebnisses
        q[deg] = coef
        # Subtrahiere coef*x^deg*d von r
        for i in range(len(d)): 
            r[i+deg]=r[i+deg]-coef*d[i]
        r=Pstandard(r); 
    q=Pstandard(q)
    return [q,r]

p=[1,2,1]; q=[1,1]
print(Pstr(p)+" / "+Pstr(q)+" = "+Pstr(Pdiv(p,q)[0])+" Rest "+Pstr(Pdiv(p,q)[1]) )
p=[0,0,0,0,-1,4]; q=[1,-1,2]
print(Pstr(p)+" / "+Pstr(q)+" = "+Pstr(Pdiv(p,q)[0])+" Rest "+Pstr(Pdiv(p,q)[1]) )
```

```python
def PggT(a:list[int], b:list[int])->list[int]: # GGT ganzzahliger Polynome 
    a = Pstandard(a); b = Pstandard(b)
    while (b != [0] and b!=[]):
        [q, r] = Pdiv(a, b) 
        [a,b]=[b,r]
    # Jetzt ist a ggT, aber Koeffizienten evtl nicht ganzzahlig
    a = Pstandard(a);
    for i in range(len(a)):
        if type(a[i])==Fraction:
            a=PSmul(a,a[i].denominator)
    # gemeinsamen Faktor entfernen:
    if len(a)>1:
        g=ggT(a[0],a[1])
        for j in range(2,len(a)):
            g=ggT(g,a[j])
            
        a=[x//g for x in a]
    return a
p=[9,-1,12,-1,3]; q=[-9,-8,-2,6,-1,3]
print("ggT("+Pstr(p)+" , "+Pstr(q)+") = "+Pstr(PggT(p,q))) 
p = [1,0,-3,2]; q = [1,0,-1]
print("ggT("+Pstr(p)+" , "+Pstr(q)+") = "+Pstr(PggT(p,q))) 
```

```python
def InterPol(xs, ys): # Interpolationspolynom nach Lagrange durch xs,ys (Listen von Zahlen)
    n=len(xs)
    coeffs= [0 for i in range(n)] 
    for j in range(n):
        xj, yj = xs[j], ys[j]
        # Lagrange-Basispolynom Lj(x)
        Lj = [1]
        nenner = 1
        for m in range(n):
            if m==j: continue
            xm=xs[m]
            nenner=nenner*(xj-xm)
            Lj = Pmul(Lj, [-xm, 1])
        # Skaliere Lj mit yj / nenner
        Lj = [(c*yj)/ Fraction(nenner) for c in Lj]
        # Summiere auf
        if len(Lj)<len(coeffs): Lj=Lj+[0]*(len(coeffs)-len(Lj))
        if len(Lj)>len(coeffs): coeffs=coeffs+[0]*(-len(coeffs)+len(Lj))
        coeffs = [coeffs[i] + Lj[i] for i in range(len(coeffs))]
    for i in range(len(coeffs)):
        if type(coeffs[i])==Fraction and coeffs[i].denominator==1:
            coeffs[i]=coeffs[i].numerator
    return coeffs
xs=[1,2,3]; ys=[5,1,5]
p=InterPol(xs, ys)
print("Interpoly durch ",xs,ys," ist",Pstr(p))
for i in range(len(xs)):
    print("    p(",xs[i],")=",Pwert(p, xs[i]),"=",ys[i])

def teiler(n):
    if n>0: return [t for t in range(1,n+1) if n%t==0]
    if n<0: return [-t for t in range(1,-n+1) if (-n)%t==0]
def PFactor(p:list[int])->list[list]:
    if not isinstance(p, list) or not all(isinstance(x, int) for x in p):
        return []
    if len(p) == 0: return []
    if len(p) == 1: return [p]
    if p[0] == 0: # man kann x faktorisieren
        return [[0,1]] + PFactor(p[1:])
    def product(list_of_lists):
        if not list_of_lists: return [[]]
        rest = product(list_of_lists[1:])
        result = []
        for item in list_of_lists[0]:
            for r in rest: result.append([item] + r)
        return result
    n = len(p) - 1  # Grad
    for d in range(1, n):  # Mögliche Grade des Faktors
        xs = list(range(1, d+2))  # d+1 Werte
        ys = [Pwert(p, x) for x in xs]
        # Alle möglichen Kombinationen g(x_i) * h(x_i) = f(x_i)
        value_sets = [teiler(y) for y in ys]
        for gs in product(value_sets):  # Alle möglichen g(x_i)
            g_poly = InterPol(xs, gs)
            if Pdeg(g_poly)<1: continue
            if g_poly is None:
                continue
            # Teste, ob g_poly wirklich Teiler von p ist
            [q, r] =  Pdiv(p, g_poly)
            if r==[]:
                if q == g_poly:  # Wiederholung erkannt
                    return [g_poly, g_poly]
                else:
                    return [g_poly] + PFactor(q)
    return [p]  # irreduzibel
p=[1,3,3,1]
pf=PFactor(p)
print(Pstr(p)+" Faktoren : "+str([Pstr(f) for f in pf]))
p=[-5,-14,-7,-13,-2,1]
pf=PFactor(p)
print(Pstr(p)+" Faktoren : "+str([Pstr(f) for f in pf]))
```

```python
import numpy as np
v = np.array([1, 2, 3])    # Vektor aus einer Liste
u = np.array([4, 5, 6])
print("u=",u,"v=",v)
print("u+v=",u+v)    # Vektoraddition
print("u-v=",u-v)    # Vektorsubtraktion
print("2*u=",2*u)    # Skalarmultiplikation
print("u*v=",u*v)    # komponentenweise Multiplikation
print("u.v=",np.dot(u,v)) # Skalarprodukt
print("uxv=",np.cross(u,v)) # Kreuzprodukt
print("|u|=",np.linalg.norm(u) ) # Länge
A = np.array([[1, 2, 0], 
              [3, 4, 7],
              [-2,3,0.5]])     # 3x3-Matrix
B = np.array([[0, 1, 0], 
              [1, 0, 1],
              [-1,0, 1]])     
print("A=",A,"B=",B)
print("A.v=",A@v) # Matrix-Vektor-Multiplikation
print("A.B=",A@B) # Matrix-Matrix-Multiplikation
print("det(A)=",np.linalg.det(A))
print("A.A^(-1)=",A@np.linalg.inv(A))
print("Eigenwerte und Eigenvektoren",list(np.linalg.eig(A)))
```

```python
# Das Folgende definiert eine einfache Turtle und öffnet die Zeichenfläche
from ipycanvas import Canvas
import math
class CanvasTurtle:
    def __init__(self, canvas, x=None, y=None, heading=0, color="black"):
        self.c = canvas
        self.width = canvas.width
        self.height = canvas.height
        self.x = x if x is not None else self.width / 2
        self.y = y if y is not None else self.height / 2
        self.heading = heading  # degrees, 0 = east
        self.pen = True
        self.color(color)
    # --- drawing control ---
    def penup(self):
        self.pen = False
    def pendown(self):
        self.pen = True
    def color(self, color):
        self.c.stroke_style = color
    def clear(self):
        self.c.clear()
        self.x = self.width/2
        self.y = self.height/2
        self.heading = 0 
        self.pen = True
        self.color("black")
    def _to_canvas(self, x, y):
        return self.width/2 + x, self.height/2 - y
    def _to_turtle(self, x, y):
        return x - self.width/2, self.height/2 - y
    # --- movement ---
    def forward(self, d):
        x2 = self.x + d * math.cos(math.radians(self.heading))
        y2 = self.y + d * math.sin(math.radians(self.heading))
        if self.pen:
            self.c.stroke_line(self.x, self.y, x2, y2)
        self.x, self.y = x2, y2
    def backward(self, d):
        self.forward(-d)
    def goto(self, x, y):
        cx, cy = self._to_canvas(x, y)
        if self.pen:
            self.c.stroke_line(self.x, self.y, cx, cy)
        self.x, self.y = cx, cy
    def getX(self): return self.x - self.width/2
    def getY(self): return self.height/2 - self.y
    # --- turning ---
    def left(self, a):
        self.heading -= a
    def right(self, a):
        self.heading += a
    def setheading(self, a):
        self.heading = a
canvas = Canvas(width=400, height=400)
t = CanvasTurtle(canvas)
```

```python
# Aufgabe 1
def Raute(a,w):
    for i in range(4):
        t.forward(a)
        if i%2==0:
            t.left(w)
        else:
            t.left(180-w)
Raute(100,45)    
display(canvas)
```

```python
def Parallelo(a,b,w):
    for i in range(4):
        if i%2==0:
            t.forward(a)
            t.left(180-w)
        else:
            t.forward(b)
            t.left(w)
t.penup(); t. goto(100,-20); t.pendown(); t.color("blue")
Parallelo(80,50,30)
```

```python
def nEck(a,n):
    for i in range(n):
        t.forward(a)
        t.left(360/n)
t.penup(); t. goto(-120,0);t.pendown(); t.color("red")
nEck(70,5)
```

```python
# Aufgabe 2
canvas = Canvas(width=400, height=400)
t = CanvasTurtle(canvas)
display(canvas)
# Abmessungen in Pixel  -b..b und -h..h
w=200; h=200
t.penup(); 
t.goto(-w,0); t.pendown(); t.goto(w,0); t.penup(); 
t.goto(0,-h); t.pendown(); t.goto(0,h);
import math
def f(x):
    return 0.9*h*math.sin(x*math.pi/w)
t.penup(); t.goto(-w,f(-w)); t.pendown(); t.color("red")
for x in range(-w,w):
    t.goto(x,f(x))
```

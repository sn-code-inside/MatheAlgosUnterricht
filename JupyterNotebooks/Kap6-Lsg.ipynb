# Kapitel6 - Lösungen

```python
# Aufgabe 1
def tau(n:int)->int:
    return sum([t for t in range(1,n+1) if n%t==0])
def perfekt(n:int)->bool:
    return tau(n)==2*n
for n in range(1,1000):
    if perfekt(n):
        print(n)
```

```python
# Aufgabe 2
def kTeiler(n:int,k:int)->bool:
    return k==len([t for t in range(1,n+1) if n%t==0])
print("3 Teiler haben",[n for n in range(1,1000) if kTeiler(n,3)]) # Quadrate von PZ
print("4 Teiler haben",[n for n in range(1,1000) if kTeiler(n,4)]) # PZ^3 oder p*q
```

```python
# Aufgabe 3
# Die vielen Fallunterscheidungen sind gar nicht nötig:
# Wenn b=1, dann ist a%b=0, d.h. in der nächsten Runde endet der Algorithmus
# Wenn a==0, dann wird im nächsten Schritt a%b=0 sein.
# a<b ist a%b=a, so dass die Rollen vertauschet sind.
def ggT(a,b): # Rekursive Definition nach Euklid
    if b==0: return a
    return ggT(b,a%b)
a=77; b=121; print("ggT(",a,",",b,")=",ggT(a,b))
```

```python
# Aufagbe 4
def QuadratischeRestLsg(a:int,n:int)->list:
    lsg=[] # Hier werden Lösungen gesammelt
    for x in range(0,n): 
        if (x*x-a)%n==0: 
            lsg.append(x) # Lösung gefunden!
    return lsg
print("Die quadratischen Reste mod 11 sind:")
for z in range(1,11):
    L= QuadratischeRestLsg(z,11)
    if len(L)>0: 
        print(z," ist quadratischer Rest, denn ",L,"^2=",z," mod 11")
```

```python
# Aufgabe 1
from vpython import *
sc1 = canvas() # Neue Szene
def Punkt(x,y,z):
   return sphere(canvas=sc1, pos=vec(x,y,z),radius=0.2,color=color.red) 
def Strecke(A,B): # zwischen Punkten A und B     
    return cylinder(canvas=sc1,pos=A.pos,axis=B.pos-A.pos,radius=0.1,color=color.blue)
# Pyramide
A=Punkt(0,0,0); B=Punkt(2,0,0); C=Punkt(2,0,2); D=Punkt(0,0,2); E=Punkt(1,2,1)
Strecke(A,B);Strecke(B,C);Strecke(C,D);Strecke(D,A);
Strecke(A,E);Strecke(B,E);Strecke(C,E);Strecke(D,E);
# Würfel
A2=Punkt(0,-2,0); B2=Punkt(2,-2,0); C2=Punkt(2,-2,2); D2=Punkt(0,-2,2);
Strecke(A2,B2);Strecke(B2,C2);Strecke(C2,D2);Strecke(D2,A2);
Strecke(A,A2);Strecke(B,B2);Strecke(C,C2);Strecke(D,D2);
```

```python
# Aufgabe 2
sc2 = canvas() 
def Punkt(x,y,z):
   return sphere(canvas=sc2, pos=vec(x,y,z),radius=0.2,color=color.red) 
P=Punkt(3,2,1) # Fester Punkt
A=Punkt(1,0,2); B=Punkt(4,-1,1) # Geradenpunkte
def Gerade(A,B): 
    # Liefert eine Parametriesierung 
    def gerade(t:float)->vec:
        return A.pos+t*(B.pos-A.pos)
    return gerade
g=Gerade(A,B)
cylinder(canvas=sc2,pos=g(-1),axis=g(2)-g(-1),radius=0.1,color=color.blue)
S=Punkt(1,0,2) # Gleiter auf Geraden
v=cylinder(canvas=sc2,pos=g(-1),axis=P.pos-g(-1),radius=0.1,color=color.blue)
t=-0.5; dt=0.1
while t<1.5:
    rate(5)      
    S.pos=g(t)
    v.pos=g(t); v.axis=P.pos-g(t)
    t+=dt
```

```python
# Aufgabe 3 - 3-Körperprobleme zeigen chaotisches Verhalten
sc5 = canvas() 
bu=10 # sowie mal groesser als masstäblich, bu= „blow up“ 
erde = sphere(canvas=sc5,pos=vec(0,0,0), 
              radius=6300000*bu,  color=color.blue)
erde.v=vec(0,0,0)
erde.m=5.9e24   # Masse der Erde
mond=sphere(canvas=sc5,pos=vec(370000000,0,0),radius=1700000*bu, color=color.red) 
mond.v = vec(0,0,370000000*6.28/(28*24*3600))# Mondgeschwindigkeit : Anfangswert
mond.m= 7.3e22 
mond2=sphere(canvas=sc5,pos=vec(220000000,220000000,0),radius=1000000*bu, color=color.red) 
mond2.v = vec(0,0,-370000000*6.28/(28*24*3600))# Mondgeschwindigkeit : Anfangswert
mond2.m= 3e22 
dt = 1800  # Zeitschritt 30 Minuten 
# Das Newton’sche Gravitationsgesetz  
def kraft(A,B): # Kraft, die B auf A ausuebt     
    gamma=6.6e-11     
    r=mag(B.pos-A.pos) # Laenge des Vektors = Abstand     
    return (B.pos-A.pos)*A.m*B.m*gamma/(r*r*r)
t=0
while t<60*24*3600: # 60 Tage
    rate(48) # 1s in Simulation entspricht 48*dt=24h     
    mond.v+= (kraft(mond,erde)+kraft(mond,mond2))/mond.m *dt
    mond2.v+= (kraft(mond2,erde)+kraft(mond2,mond))/mond2.m *dt
    erde.v+= (kraft(erde,mond)+kraft(erde,mond2))/erde.m *dt     
    mond.pos+= mond.v *dt     
    mond2.pos+= mond2.v *dt     
    erde.pos+= erde.v *dt
    t+=dt
```

```python
# Aufgabe 2
class Bruch:
    def __init__(self, z, n): # z,n = Zähler,Nenner
        self.z = z
        self.n = n
    def __repr__(self): # Ausgabe 
        return f"{self.z}/{self.n}"
    def __add__(self, other): # Addition
        return Bruch(self.z*other.n+self.n*other.z, self.n*other.n)
    def __sub__(self, other): # Subtraktion
        return Bruch(self.z*other.n-self.n*other.z, self.n *other.n)
    def __mul__(self, other): # Multiplikation
        return Bruch(self.z*other.z, self.n*other.n)
    def __truediv__(self, other): # Division
        return Bruch(self.z*other.n, self.n*other.z)
b1 = Bruch(3, 4)     # 3/4
b2 = Bruch(7, 6)    # 7/6
print("b1 =", b1, "b2 =", b2)
print("b1 + b2 =", b1 + b2)
print("b1 - b2 =", b1 - b2)
print("b1 * b2 =", b1 * b2)
print("b1 / b2 =", b1 / b2)
```

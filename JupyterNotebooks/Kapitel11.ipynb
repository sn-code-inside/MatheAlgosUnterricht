```python
sc2 = canvas() # Neue Szene
def Punkt(x,y,z):
   return sphere(canvas=sc2, pos=vec(x,y,z),radius=0.3,color=color.red) 
def Strecke(A,B): # zwischen Punkten A und B     
    return cylinder(canvas=sc2,pos=A.pos,axis=B.pos-A.pos,radius=0.1,color=color.blue)
n=20 
Punkte=[] 
from random import randint
for i in range(n):     
    x=randint(-10,10)     
    y=randint(-10,10)     
    z=randint(-10,10)
    Punkte.append(Punkt(x,y,z)) 
for A in Punkte:
   for B in Punkte:   
       Strecke(A,B) 
```

```python
sc3 = canvas() # Neue Szene
import math
Punkte=[] 
for alpha in range(3*360):
    x=5*math.sin(alpha/180.0*math.pi)     
    y=5*math.cos(alpha/180.0*math.pi)     
    z=3*alpha/360.0-5
    Punkte.append(vec(x,y,z))
curve(canvas=sc3, pos=Punkte, radius=0.05) 
# Jetzt noch einen animierten Punkt  
ball=sphere(canvas=sc3, radius=0.5,color=color.red) 
for p in Punkte:     
    ball.pos=p     
    rate(20) 
```

```python
sc4 = canvas() 
boden = box(canvas=sc4, size=vec(8,0.5,8), color=color.blue) 
ball = sphere(canvas=sc4, pos=vec(0,4,0), color=color.red) 
ball.v = vec(0,-1,0)  # Ballgeschwindigkeit: Anfangswert 
ball.m=5  # Ballmasse 
unten=vec(0,-1,0)     # Richtung der Schwerkraft 
g=9.81      # Erdbeschleunigung 
t=0; dt = 0.01                # Zeitschritt
while t<10:     
    rate(100)  
    t+=dt
    kraft=g*unten*ball.m
    ball.v+= kraft/ball.m*dt
    ball.pos+= ball.v*dt     
    if ball.pos.y < 1: ball.v.y = -ball.v.y # Springen
```

```python
sc5 = canvas() # Neue Szene
bu=10 # sowie mal groesser als masstäblich, bu= „blow up“ 
erde = sphere(canvas=sc5,pos=vec(0,0,0), 
              radius=6300000*bu,  color=color.blue)
erde.v=vec(0,0,0)
erde.m=5.9e24   # Masse der Erde
mond=sphere(canvas=sc5,pos=vec(370000000,0,0),radius=1700000*bu, color=color.red) 
mond.v = vec(0,0,370000000*6.28/(28*24*3600))# Mondgeschwindigkeit : Anfangswert
mond.m= 7.3e22 
dt = 1800  # Zeitschritt 30 Minuten 
# Das Newton’sche Gravitationsgesetz  
def kraft(A,B): # Kraft, die B auf A ausübt     
    gamma=6.6e-11     
    r=mag(B.pos-A.pos) # Länge des Vektors = Abstand     
    return (B.pos-A.pos)*A.m*B.m*gamma/(r*r*r)
t=0
while t<30*24*2*dt:
    rate(48) # 1s in Simulation entspricht 48*dt=24h     
    mond.v+= kraft(mond,erde)/mond.m*dt
    erde.v+= kraft(erde,mond)/erde.m *dt     
    mond.pos+= mond.v *dt     
    erde.pos+= erde.v *dt
    t+=dt
```

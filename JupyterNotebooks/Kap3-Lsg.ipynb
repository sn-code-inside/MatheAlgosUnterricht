```python
# Aufgabe 1
import math
def FVol(d,h,h2): # Volumen einer Flache mit Durchmesser d, Höhe des zyblindsichen Teils h und Höhe des Kegels h2
    VZyl=math.pi*h*(d/2)**2
    VKegel=1/3*math.pi*h2*(d/2)**2
    return VZyl+VKegel
FVol(8,12,8)    
```

```python
# Aufgabe 2
def PotenzSumme(n,expo=2):
    s=0
    for i in range(1,n+1):
        s+=i**expo
    return s
def PotenzSummeR(n,expo=2):
    if n==0: return 0
    return n**expo + PotenzSummeR(n-1,expo)
print("Quadratsumme 1..10",PotenzSumme(10),PotenzSummeR(10))
print("Kubiksumme 1..10",PotenzSumme(10,expo=3),PotenzSummeR(10,expo=3))
```

```python
# Aufgabe 3 - Bei diesem Beispiel ist die rekursive Version viel eleganter, aber bei größeren Zahlen auch viel langsamer
def Fibo(n): # iterativ
    a=1
    b=1
    for i in range(n-1):
        nächste=a+b
        a=b
        b=nächste
    return b
def FiboR(n): # rekursiv
    if n==1 or n==0: return 1
    return FiboR(n-1)+FiboR(n-2)
n=10; print("Fib(",n,")=", Fibo(n),"=", FiboR(n))
```

```python
# Aufgabe 4
def GSi(n):
    s=0
    for i in range(0,n+1):
        s+=q**i
    return s
def GSr(n):
    if n==0: return 1
    return q**n+GSr(n-1)
def GSe(n): 
    return (1-q**(n+1))/(1-q)
q=0.9
for e in range(1,4):
    n=10**e
    print(n,GSi(n),GSr(n),GSe(n))
```

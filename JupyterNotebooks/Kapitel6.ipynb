```python
def teilt(teiler:int, n:int)->bool: 
    return n%teiler==0  
def teiler(n:int)->list[int]: # Gibt eine Liste aller Teiler 
    return [t for t in range(1,n+1) if teilt(t,n)]
print("Ist 5 Teiler von 37? ", teilt(5,37)) 
print("Die Teiler von 24 sind: ", teiler(24))
```

```python
def prim(n:int)->bool: 
    return len(teiler(n))==2
for i in range(1,15):
    print(i,prim(i))
```

```python
def primRange(a:int,b:int)->list[int]: 
   return [p for p in range(a,b) if prim(p)]
primRange(20,40)
```

```python
def Sieb(n:int)->list[int]:       
    primzahl=[True]*(n+1) 
    primzahl[0]=False   # 0 und 1 sind keine Primzahlen     
    primzahl[1]=False     
    p=2               # 2 ist die erste PZ     
    while True:
        i=2
        while i*p<=n: # Alle Vielfachen von p streichen            
            primzahl[i*p]=False
            i+=1
        q=-1 # Speicher für evtl. weitere Primzahl                
        for z in range(p+1,n+1): # Naechste PZ suchen
            if primzahl[z]:
                q=z
                break 
        if q==-1: # Keine weiteren Primzahlen, fertig
            break
        p=q  # Die gefundene Zahl ist die naechste PZ  
    return [p for p in range(2,n+1) if primzahl[p]]
Sieb(100)
```

```python
def ggt(a:int,b:int)->int: 
   ta=teiler(a) # Teilerliste von a      
   ta.reverse() # umsortiert, grosse zuerst
   tb=teiler(b)     
   for t in ta: # Teiler von a von gross nach klein 
       if t in tb: 
           return t     
   return 1
ggt(24,36)
```

```python
def ggT(a,b): # Rekursive Definition nach Euklid  
    if a==b: 
        return a
    if a==1 or b==1: 
        return 1
    if a==0: 
        return b 
    if b==0: 
        return a
    if a>b: 
        return ggT(a%b,b)  # Es ginge auch: ggT(a-b,b)
    if a<b: return ggT(a,b%a)
ggT(24,36)
```

```python
# Variante des Algorithmus mit Schleife und Listenzuweisung
def ggT(a,b): 
    while b !=0:
        [q, r] = [a//b, a%b] 
        [a,b]=[b,r]
    return a
ggT(10,15)
```

```python
def kgV(a,b): 
    return a*b//ggT(a,b)
kgV(12,14)
```

```python
def primZerlegung(n:int)->list[int]: 
    if n<2: 
        return []
    if prim(n):
        return [n]
    for t in range(2,n):
        if teilt(t,n): 
            return [t]+primZerlegung(n//t)
primZerlegung(5133)
```

```python
def phi(n:int)->int: 
    return len([i for i in range(1,n) if ggT(i,n)==1])
phi(5*7)
```

```python
def EggT(a:int, b:int)->list: # Liefert [g,x,y] mit g=x*a+y*b=ggT(a,b)
    if b>a:
        L=EggT(b,a)
        return [L[0],L[2],L[1]] # Umdrehen, wenn b>a
    if a==b: 
        return [a,1,0]  # denn ggT(a,a)=a=1*a+0*b    
    if b==0: 
        return [a,1,0]  # denn ggT(a,0)=a=1*a+0*b
    q=a//b
    L=EggT(b,a%b) 
    return [L[0],L[2],L[1]-q*L[2]]
print(EggT(152,176))
7*152-6*176
```

```python
def diophant(a:int,b:int,c:int)->list: # Löst a*x+b*y=c     
    g=ggT(a,b)     
    if not(teilt(g,c)): return []
    L=EggT(a//g,b//g)
    return [c//g*L[1],c//g*L[2]]
print(diophant(152,176,32))
152*28+176*(-24)==32
```

```python
def multInv(a:int,n:int)->int:
    L=diophant(a,n,1)
    if L==[]:
        raise Exception("Mult. In.v existiert nicht")
    return L[0]%n
a=13; n=45
i=multInv(a,n)
print("Inverses von a=",a,"mod n=",n,"ist i=",i,"denn a*i%n=",a*i%n) 
```

```python
def Text2List(t:str)->list:
    return [ord(c) for c in t]
print(Text2List("Augsburg"))
def List2Text(L:list)->str:
    s=""
    for x in L:
        s=s+chr(x)
    return s
print(List2Text([65,66,67]))
def Caesar(t:str,schlüssel:int)->str:
    return List2Text([x+schlüssel for x in Text2List(t)])
geheimtext=Caesar("AUGSBURG",2)
print("verschlüsselt:",geheimtext)
print("entschlüsselt:",Caesar(geheimtext,-2))
```

```python
def nextPrime(n): # gibt naechste Primzahl nach n     
    kandidat=n+1     
    while not(prim(kandidat)):
        if kandidat%2==1: 
            kandidat+=2
        else: 
            kandidat+=1     
    return kandidat
print(nextPrime(25))
```

```python
import random
def ErzeugeRSASchlüssel(n0:int)-> list: # n0 ist untere Grenze der Größe der PZ
    p=nextPrime(random.randrange(n0,2*n0))
    q=nextPrime(random.randrange(n0,2*n0))
    n=p*q
    phi=(p-1)*(q-1)
    e=random.randrange(n//2,n) 
    while not(ggT(e,phi)==1): # Suche e teilerfremd zu phi
        e+=1
    d=multInv(e,phi)
    return [[e,n],d]
S1=ErzeugeRSASchlüssel(1000) # Schlüssel von Person 1
print("Öffentlicher Schlüssel von 1:",S1[0])
print("Geheimer Schlüssel von 1:",S1[1])
S2=ErzeugeRSASchlüssel(1000) # Schlüssel von Person 2
print("Öffentlicher Schlüssel von 2:",S2[0])
print("Geheimer Schlüssel von 2:",S2[1])
S2oeff=S2[0]
```

```python
def expmod(a,expo,n): # berechnet (a**expo)%n relativ schnell
    if expo==0: 
        return 1
    if expo==1: 
        return a%n
    if expo%2==1: return (a*expmod(a,expo-1,n))%n
    if expo%2==0: return (expmod(a,expo//2,n)**2)%n
def Sende(a:int,Soeff:list)->int: # Sende Nachricht an Empfänger mit öffentlichem Schlüssel S
    e=Soeff[0]; n=Soeff[1];  
    if a>=n:
        raise Exception("Man kann nur Nachrichten-Zahlen kleiner n senden")
    return expmod(a,e,n)
def Entschlüssele(c:int,S:list)->int:
    Soeff=S[0]; e=Soeff[0]; n=Soeff[1]; d=S[1]
    return expmod(c,d,n)
a=49714 # Soll an Person 2 gesendet werden
c=Sende(a,S2oeff)
print("c=",c,"kann öffentlich an Person 2 übermittelt werden")
print("Empfänger S2 entschlüsselt mit seinem d und findet:", Entschlüssele(c,S2))   
```

```python
import math
def FaktorFermat(n):
    if n%2 == 0: return [2, n//2] 
    a = int(math.sqrt(n))+1
    while True:
        b2 = a*a-n
        b=int(math.sqrt(b2))
        if b*b == b2: # Ist b2 ein Quadrat?
            return [a-b, a+b]
        a+= 1
print(FaktorFermat(105))
print(FaktorFermat(1076859))
```

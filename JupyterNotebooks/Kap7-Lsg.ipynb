```python
# Aufgabe 1
import math
n=6; s=1;
for j in range(15):
    n=2*n; 
    h=1-math.sqrt(1-(s/2)**2)
    s=math.sqrt(2-2*math.sqrt(1-(s/2)**2))
    S=s*1/(1-h) # zentrische Streckung
    p=n/2*s; P=n/2*S;
    print(n,":",p,P,P-p)
```

```python
# Polynome - Wiederholung der Definitionen aus dem Buch
from fractions import Fraction
def Pwert(p:list,x): # Wert der Polynomfunktion
    return sum([p[k]*x**k for k in range(len(p))])
def Pstr(p:list,var="x")->str:
    if p==[]: return "0"
    s=""
    for i in range(len(p)-1,-1,-1):
        if p[i]==0:
            continue
        if i==0:
            s=s+"+"+str(p[i])
        if i==1:
            s=s+"+"+str(p[i])+"*"+var
        if i>1:
            if i==len(p)-1:
                s=str(p[i])+"*"+var+"^"+str(i)
            else: 
                s=s+"+"+str(p[i])+"*"+var+"^"+str(i)
    return s
def Pstandard(p:list)->list:
    while len(p)>0 and p[-1]==0: 
        p=p[0:len(p)-1]
    return(p)
def Pdeg(p:list)->int: # Grad
    return len(Pstandard(p))-1
def Padd(a:list,b:list)->list: # Addition
    while len(a)<len(b):
        a=a+[0] 
    while len(b)<len(a):
        b=b+[0]         
    return Pstandard([a[i]+b[i] for i in range(len(a))])
def Psub(a:list,b:list)->list: 
    while len(a)<len(b): # Subtraktion
        a=a+[0] 
    while len(b)<len(a):
        b=b+[0]         
    return Pstandard([a[i]-b[i] for i in range(len(a))])
def PSmul(p:list, s)->list: # Multiplikation mit Zahl s
    return Pstandard([s*x for x in p]) 
def Pmul(a:list,b:list)->list:
    res=[0]*(1+(len(a)-1)+(len(b)-1)) 
    # Ergebnis wird Polynom vom Grad(a)+Grad(b)
    for i in range(len(a)):
        for j in range(len(b)):
            res[i+j]+=a[i]*b[j]
    return Pstandard(res)
def Pdiv(dividend:list, divisor:list)->list[list]:
    def quot(a,b): # Division, ggf als Bruch
        if type(a)==int and type(b)==int: 
            if a%b==0: 
                return a//b
            else:
                return Fraction(a,b)
        return a/b
    # Kopien anlegen und führende Nullen entfernen
    r=Pstandard(dividend[:]); d=Pstandard(divisor[:])
    if len(d)==0 or (len(d) == 1 and d[0] == 0):
        raise ZeroDivisionError("Division durch das Nullpolynom")
    m = len(r)-1 # Grad dividend
    n = len(d)-1 # Grad divisor
    if m<n: return [[0],r]
    q=[0]*(m-n+1) # Liste passender länge für den Quotienten
    while len(r)-1>=n:
        coef=quot(r[-1],d[-1])
        deg =(len(r)-1)-n # Grad des Ergebnisses
        q[deg] = coef
        # Subtrahiere coef*x^deg*d von r
        for i in range(len(d)): 
            r[i+deg]=r[i+deg]-coef*d[i]
        r=Pstandard(r); 
    q=Pstandard(q)
    return [q,r]
def ggT(a,b):  
    if b==0: 
        return a
    return ggT(b,a%b)
def PggT(a:list[int], b:list[int])->list[int]: # GGT ganzzahliger Polynome 
    a = Pstandard(a); b = Pstandard(b)
    while (b != [0] and b!=[]):
        [q, r] = Pdiv(a, b) 
        [a,b]=[b,r]
    # Jetzt ist a ggT, aberKoeffizienten evtl nicht ganzzahlig
    a = Pstandard(a);
    for i in range(len(a)):
        if type(a[i])==Fraction:
            a=PSmul(a,a[i].denominator)
    # gemeinsamen Faktor entfernen:
    if len(a)>1:
        g=ggT(a[0],a[1])
        for j in range(2,len(a)):
            g=ggT(g,a[j])      
        a=[x//g for x in a]
    return a
```

```python
# Aufgabe 3 : [z,n] ist rationaler Term mit Zähler und Nenner als Polynome
def Rstr(r:list[list])->str:
    return "("+Pstr(r[0])+") / ("+Pstr(r[1])+")"
def Rkurz(r:list[list])->list[list]:
    # Vereinfacht und kürzt
    [z,n]=r
    g=PggT(z,n)
    return [Pdiv(z,g)[0], Pdiv(n,g)[0]]
r=[[1,3,3,1],[1,2,1]]
print("r=",Rstr(r),"=",Rstr(Rkurz(r)))
r=[[-15,10,-3,2],[6,-7,2,-15,10]]
print("r=",Rstr(r),"=",Rstr(Rkurz(r)))
def Radd(r1:list[list],r2:list[list])->list[list]:
    [z1,n1]=r1; [z2,n2]=r2 
    return Rkurz([Padd(Pmul(z1,n2),Pmul(z2,n1)),Pmul(n1,n2)])
def Rsub(r1:list[list],r2:list[list])->list[list]:
    [z1,n1]=r1; [z2,n2]=r2 
    return Rkurz([Psub(Pmul(z1,n2),Pmul(z2,n1)),Pmul(n1,n2)])
def Rmul(r1:list[list],r2:list[list])->list[list]:
    [z1,n1]=r1; [z2,n2]=r2 
    return Rkurz([Pmul(z1,z2),Pmul(n1,n2)])
def Rdiv(r1:list[list],r2:list[list])->list[list]:
    [z1,n1]=r1; [z2,n2]=r2 
    return Rkurz([Pmul(z1,n2),Pmul(n1,z2)])
def Rableitung(r:list[list])->list[list]:
    [z,n]=r
    return Rkurz([Psub(Pmul(Pableitung(z),n),Pmul(z,Pableitung(n))) , Pmul(n,n)])
print(Rstr(r)+" abgeleitet: ", Rstr(Rableitung(r)))
```

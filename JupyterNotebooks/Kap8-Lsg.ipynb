```python
# Aufgabe 1
# Die Additionstheoreme liefern 
# tan(2x)=sin(2x)/cos(2x)=2sin(x)cos(x)/(cos(x)^2-sin(x)^2)=2tan(x)/(1-tan(x)^2)
def Tan(x, eps=1e-9):
    if abs(x)<eps:
        return x
    t2=Tan(x/2)
    return 2*t2/(1-t2**2)
import math
print("tan(1.2)=",Tan(1.2),"=",math.tan(1.2))
```

```python
# Aufgabe 2
def heronM(a:float,m:int,x0=1,eps=1e-7): # m-te Wurzel
    if abs(x0**m-a)<eps:
        return x0
    return heronM(a,m,(x0+a/x0**(m-1))/2,eps)
print("3. Wurzel aus 15=",heronM(15,3))    
```

```python
# Aufgabe 3
def Log(a,b,steps=50,eps=1e-5): # Log(a) zur Basis b
    if a<=0 or b<=0 or b==1:
        raise Exception("Log nur für positive Zahlen")
    if a<1: 
        return -Log(1/a,b,steps-1)
    if abs(a-b)<eps:
        return 1
    if abs(a-1)<eps:
        return 0
    if steps<0:
        print("Warnung: Genauigkeit nicht erreicht")
        return a-1
    if b<=a:
        return 1+Log(a/b,b,steps-1)
    else:
        return 1/2*Log(a**2,b,steps-1)
print(Log(15,2),"=?",math.log(15)/math.log(2))
```

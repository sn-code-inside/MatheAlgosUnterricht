```python
def alsMenge(L):
    M=[]
    for x in L:
        if not(x in M):
            M.append(x)
    return M
alsMenge([5,3,4,1,4,6,1])
```

```python
def Vereinigungsmenge(A,B):
    return alsMenge(A+B)
print(Vereinigungsmenge([1,2,3,6],[6,3,4,5]))
def Schnittmenge(A,B):
    return [x for x in A if x in B]
print(Schnittmenge([1,2,3,6],[6,3,4,5]))
def Komplement(A,B):
    return [x for x in A if not(x in B)]
print(Komplement([1,2,3,6],[6,3,4,5]))
```

```python
def varianz(liste):
   m=mittelwert(liste)     
   return mittelwert([(x-m)**2 for x in liste]) 
print(mittelwert(L),", ", varianz(L)) 
```

```python
import math
def standardabweichung(L):
    return math.sqrt(varianz(L))
def median(L):     
    sortiert=sorted(L)
    n=len(L)
    if n%2==1: # ungerade Anzahl
        return sortiert[(n-1)//2] 
    else: return (sortiert[n//2-1]+sortiert[n//2])/2
median(L)
```

```python
def standardisiert(liste): # Liste mit Mittel 0, SD 1
    m=mittelwert(liste) 
    s=math.sqrt(varianz(liste)) # Standardabweichung
    return [ (x-m)/s for x in liste ]
Ls=standardisiert(L)
[mittelwert(Ls),standardabweichung(Ls)]
```

```python
def korrel(A,B):     
    if len(A)!=len(B):
        print("Fehler in korrel: Ungleiche Längen")
        return 0
    As=standardisiert(A)     
    Bs=standardisiert(B) 
    return sum([As[i]*Bs[i] for i in range(len(A))])/len(A)
schuhe= [41,  42, 43, 44, 35, 36] 
körper=[179,185,180,177,166,165] 
print("Korrelation(",schuhe,",",körper,")=",korrel(schuhe,körper))
```

```python
def kaufenBisAlle3():
    k=0 # Anzahl der gekauften Packungen
    habschon=[] # Figuren, die man schon hat
    while len(habschon)<3:
        k+=1 # Kaufe eine weitere Packung
        z=random.randrange(1,4) #darin befindet sich zufällig z 
        if not(z in habschon):
            habschon.append(z)
    return k
for i in range(5):
    print(kaufenBisAlle3())
print("Erwartungswert",mittelwert([kaufenBisAlle3() for i in range(1000)]))
```

```python
def geburt(): # 0=Mädchen, 1=Junge     
    if random.random()<=0.51: 
        return 1     
    else: 
        return 0
def NJungen(n): # Anzahl der Jungen bei n simulierten Geburten
    return sum([ geburt() for i in range(n)]) 
# Jetzt 10000 mal 120 Geburten simulieren,  # zählen wenn 68 oder mehr Jungs 
m=10000; n=120; g=68; k=0 
for i in range(m):  # m mal n Geburten simulieren     
    j=NJungen(n)
    if j>=g: 
        k=k+1
p=k/m 
print("Wahrscheinlichkeit für mindestens ",g," Jungs bei",n," Geburten: p=",p,) 
if p<0.05: print("signifikant auffällig") 
else: print("nicht auffällig")
```

```python
# Test zwischen Gruppen
A = [12, 11, 13, 10, 14, 16]   # Stichprobe A
B = [15, 14, 16, 13, 15, 12, 15]   # Stichprobe B
def MWDiff(A,B): # Mittelwertdifferenz
    return mittelwert(B)-mittelwert(A)   
print("Mittelwertdifferenz ",MWDiff(A,B))
C=A+B
cnt = 0
N = 10000
for j in range(N):
    random.shuffle(C)
    Asimu = C[:len(A)]
    Bsimu = C[len(A):]
    if MWDiff(Asimu, Bsimu) >= MWDiff(A,B):
        cnt += 1
print("Wenn H0 richtig, hat diese Mittelwertsdifferenz einen p-Wert von ",cnt/N)
```

```python
# Erzeugungung von standard-NV Zufallszahlen
def U(): # Uniform E(U)=0, V(U)=1
    return (random.random()-0.5)*math.sqrt(12)
sample=[U() for i in range(1000)]
print("U",[mittelwert(sample),standardabweichung(sample),min(sample),max(sample)])
# Standardnormalverteilt
def SNV(n=1000):
    return sum([U() for i in range(n)])/math.sqrt(n)
sample=[SNV() for i in range(1000)]
print("SNV",[mittelwert(sample),standardabweichung(sample),min(sample),max(sample)])
def NV(mu,sigma):
    return mu+sigma*SNV()
sample=[NV(3,5) for i in range(1000)]
print("NV(3,5)",[mittelwert(sample),standardabweichung(sample),min(sample),max(sample)])
```

```python
# Monte-Carlo Methode für pi
n=100000 # 100000 Tropfen 
trefferZahl=0 
for i in range(n):     
    x=random.random()     
    y=random.random()     
    if x**2+y**2<1: 
        trefferZahl+=1 
p=trefferZahl/n
print("pi ist ca.: ", 4*p)
```

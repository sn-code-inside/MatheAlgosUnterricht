```python
# Aufgabe 5
n=15
# 1. Möglichkeit
s=0; k=1
for i in range((n-1)//2):
    s+=k # Kurzform von s=s+k
    k+=2
print("1. Möglichkeit:",s)
s=0; 
for i in range(1,n,2): # Nutzt range mit Schrittweite
    s+=i
print("2. Möglichkeit:",s)    
```

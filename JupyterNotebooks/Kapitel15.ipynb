```python
class KomplexeZahl:
    def __init__(self, re: float, im: float = 0.0):
        self.re = re
        self.im = im
    def __repr__(self): # Ausgabe als Zahl
        vorzeichen = "+" if self.im >= 0 else "-"
        return f"{self.re} {vorzeichen} {abs(self.im)}i"
    def __add__(self, other): # Addition
        return KomplexeZahl(self.re + other.re, self.im + other.im)
    def __sub__(self, other): # Subtraktion
        return KomplexeZahl(self.re - other.re, self.im - other.im)
    def __mul__(self, other): # Multiplikation
        r = self.re * other.re - self.im * other.im
        i = self.re * other.im + self.im * other.re
        return KomplexeZahl(r, i)
    def __truediv__(self, other): # Division
        nenner = other.re**2 + other.im**2
        if nenner == 0:
            raise ZeroDivisionError("Division durch 0")
        r = (self.re * other.re + self.im * other.im) / nenner
        i = (self.im * other.re - self.re * other.im) / nenner
        return KomplexeZahl(r, i)
    def konjugiert(self): # liefert konjugierte Zahl
        return KomplexeZahl(self.re, -self.im)
    def konjugiere(self): # verändert self
        self.im = -self.im
        return self
z1 = KomplexeZahl(3, 4)     # 3 + 4i
z2 = KomplexeZahl(1, -2)    # 1 - 2i
print("z1 =", z1)
print("z2 =", z2)
print("z1 + z2 =", z1 + z2)
print("z1 - z2 =", z1 - z2)
print("z1 * z2 =", z1 * z2)
print("z1 / z2 =", z1 / z2)
print("konjugiert von z1 =", z1.konjugiert())
print("z1 ist unverändert",z1)
z1.konjugiere()
print("z1 nach konjugiere() =", z1)
```

```python
# Veriable Argumentzahl
def addiere(a,b,*rest):
    summe=a+b
    for x in rest:
        summe+=x
    return summe
addiere(1,2,3,4,5)
```

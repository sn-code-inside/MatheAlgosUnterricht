```python
# Aufgabe 1
def mymax(L:list)->float: # Rekursive Variante
    if len(L)==1: return L[0] # Achtung mymax kann nicht mit [] umgehen
    if L[0]>mymax(L[1:]): 
        return L[0]
    else: return mymax(L[1:])
def mysum(L:list)->float: # Rekursive Variante
    if len(L)==0: return 0
    return L[0]+mysum(L[1:])
a=[1,5,3,2]
print(mymax(a))
print(mysum(a))
```

```python
def mymax(L:list)->float: # Iterative Variante
    m=L[0]
    for x in L:
        if x>m:
            m=x
    return m
def mysum(L:list)->float: 
    s=0
    for x in L:
        s+=x
    return s
a=[1,5,3,2]
print(mymax(a))
print(mysum(a))
```

```python
# Aufgabe 4
def fak(n:int)->int:
    if n<2: return 1
    else: return n*fak(n-1)
def bino(n:int,k:int)->float: return fak(n)/(fak(n-k)*fak(k))
def binodist(n:int,k:int,p:float)->float: return p**k*(1-p)**(n-k)*bino(n,k)
def kumbinodist(n:int,k:int,p:float)->float: 
    return sum([p**j*(1-p)**(n-j)*bino(n,j) for j in range(k+1)])
print("Test, n=10, p=0.5")
for k in range(10+1): print([k,bino(10,k),binodist(10,k,0.5),kumbinodist(10,k,0.5)])
```

```python
# Aufgabe 5
def vadd(a:list, b:list)->list:
    if len(a)!=len(b): # Die beiden Zeilen könnte man sich natürlich sparen, aber sicher ist sicher...
        raise Exception("vadd: Vektoren müssen gleiche Dim. haben")
    return [a[i]+b[i] for i in range(len(a))]
def vsub(a:list, b:list)->list:
    if len(a)!=len(b):
        raise Exception("vsub: Vektoren müssen gleiche Dim. haben")
    return [a[i]-b[i] for i in range(len(a))]
def smul(x:float, b:list)->list:
    return [x*b[i] for i in range(len(b))]
def skalarp(a:list, b:list)->float:
    if len(a)!=len(b):
        raise Exception("Skalarprodukt: Vektoren müssen gleiche Dim. haben")
    return sum([a[i]*b[i] for i in range(len(a))])
def betrag(v:list)->float:
    return math.sqrt(skalarp(v,v))
u=[2,1,4]; v=[-2,7,5]
print("u+v=",vadd(u,v),"u-v=",vsub(u,v),"u+3*v=",vadd(u,smul(3,v)),"u.v=",skalarp(u,v),"|u|=",betrag(u))
```

```python
# Aufgabe 6
import math
def shape(A:list)->list:
    """Gibt [m,n] zurück für eine m×n-Matrix A (Liste von Zeilen)."""
    if not isinstance(A, list) or len(A) == 0:
        raise Exception("shape: Matrix muss eine nicht-leere Liste sein")
    if not all(isinstance(row, list) for row in A):
        raise Exception("shape: Matrix muss Liste von Listen sein")
    n = len(A[0])
    if n == 0:
        raise Exception("shape: Zeilen dürfen nicht leer sein")
    if any(len(row) != n for row in A):
        raise Exception("shape: Alle Zeilen müssen gleich lang sein")
    return [len(A), n]

def copy_matrix(A:list)->list: return [row[:] for row in A]

# Matrixaddition und -subtraktion 
def madd(A:list, B:list)->list:
    if shape(A) != shape(B):
        raise Exception("madd: Matrizen müssen gleiche Dimension haben")
    [mA,nA] = shape(A)
    return [[A[i][j] + B[i][j] for j in range(nA)] for i in range(mA)]
def msub(A:list, B:list)->list:
    if shape(A) != shape(B):
        raise Exception("msub: Matrizen müssen gleiche Dimension haben")
    [mA,nA] = shape(A)
    return [[A[i][j] - B[i][j] for j in range(nA)] for i in range(mA)]
    
# Matrixmultiplikation
def mmul(A:list, B:list)->list:
    [mA,nA] = shape(A)
    [mB,nB] = shape(B)
    if nA != mB:
        raise Exception("mmul: Dimension passt nicht (A ist m×n, B ist n×k nötig)")
    return [[sum(A[i][k] * B[k][j] for k in range(nA)) for j in range(nB)]
            for i in range(mA)]

# Multiplikation Matrix*Vektor 
def mvec(A:list, x:list)->list:
    [mA,nA] = shape(A)
    if len(x) != nA:
        raise Exception("mvec: Dimension passt nicht")
    return [sum(A[i][k] * x[k] for k in range(nA)) for i in range(mA)]

# Gauß-Algorithmus (Zeilenstufenform)
def gauss_elim(A:list, eps:float=1e-12)->list:
    # Veränderbare Kopie:
    M = [[float(x) for x in row] for row in copy_matrix(A)]
    [m,n] = shape(M)
    row = 0
    for col in range(n):
        # Pivot suchen: größte |Eintrag| ab row
        pivot = None; best = 0.0
        for r in range(row, m):
            val = abs(M[r][col])
            if val > best:
                best = val; pivot = r
        if pivot is None or best < eps:
            continue  # keine brauchbare Pivot-Spalte
        # Pivotzeile nach oben tauschen
        if pivot != row:
            M[row], M[pivot] = M[pivot], M[row]
        # Unterhalb eliminieren
        for r in range(row + 1, m):
            if abs(M[r][col]) < eps: continue
            factor = M[r][col] / M[row][col]
            # M[r] = M[r] - factor * M[row]
            for c in range(col, n):
                M[r][c] -= factor * M[row][c]
        row += 1
        if row == m: break
    return M


# Determinante
def minor(M,i,j): # Löscht Zeile i und Spalte j
    M=[ zeile[:j]+zeile[(j+1):] for zeile in M]
    return M[:i]+M[(i+1):]
def det(M): # Mit Entwicklungsatz berechnen
    [n,m]=shape(M)
    if n!=m: raise Exception("Det erforder quadrat. Matrix")
    if n==1: return M[0][0]  # 1x1 Matrix
    d=0
    for j in range(n):
        d+=(-1)**j*M[j][0]*det(minor(M,j,0))
    return d
def replace_col(A, col, b):
    # Ersetzt Spalte col in A durch den Vektor b
    [n, m] = shape(A)
    if len(b) != n:
        raise Exception("replace_col: b hat falsche Länge")
    return [
        A[i][:col] + [b[i]] + A[i][col+1:] for i in range(n) ]

def cramer_solve(A, b):
    # Löst A x = b mit der Cramer-Regel.
    [n, m] = shape(A)
    if n != m:
        raise Exception("cramer_solve: A muss quadratisch sein")
    if len(b) != n:
        raise Exception("cramer_solve: b muss passende Länge haben")
    dA = det(A)
    if dA == 0:
        raise Exception("cramer_solve: det(A)=0, keine eindeutige Lösung")
    x = []
    for i in range(n):
        Ai = replace_col(A, i, b)
        x.append(det(Ai) / dA)
    return x

# Mini-Test
A = [
    [2, 1, 4],
    [-2, 7, 5],
    [1, 0, 3]
]
B = [
    [1, 2, 0],
    [0, 1, 1],
    [2, 0, 1]
]
b = [1, 2, 3]

print("A+B =", madd(A, B))
print("A-B =", msub(A, B))
print("A*B =", mmul(A, B))
print("det(A) =", det(A))
print("Zeilenstufenform(A) =", gauss_elim(A))
print("Lösung von A x = b:", cramer_solve(A, b))
```

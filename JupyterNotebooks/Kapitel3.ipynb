```python
def Nsumme(n):     
    summe=0
    for i in range(1,n+1):
        summe=summe+i
    return summe
Nsumme(100)
```

```python
import math
def piApprox(n): # n ist die Zahl der Streifen     
    fläche=0.0     
    delta=1/n # Streifenbreite     
    for i in range(n):         
        x=i*delta         
        y=math.sqrt(1.0-x**2)         
        fläche+=y*delta     
    return 4*fläche # Bisher nur Viertelkreis berechnet
piApprox(1000)
```

```python
for i in range (1,2002, 100):   
    print(i, math.pi-piApprox(i))
```

```python
def piApprox(n): # n ist die Zahl der Streifen    # Werte in der Mitte des Streifens  
    fläche=0.0     
    delta=1/n # Streifenbreite     
    for i in range(n):         
        x=(i+0.5)*delta         
        y=math.sqrt(1.0-x**2)         
        fläche+=y*delta     
    return 4*fläche # Bisher nur Viertelkreis berechnet
piApprox(1000)
```

```python
def fakultät(n):
    fakul=1   
    for i in range(1,n+1):
        fakul=fakul*i     
    return fakul 
fakultät(5)
```

```python
def potenz(a,n): # berechnet a hoch n   
    ergebnis=1
    for i in range(n):
        ergebnis = ergebnis * a
    return ergebnis 
potenz(5,3)
```

```python
def bino(n,k): # Variante mit direkter Berechnung
    ergebnis=1
    for i in range(k+1,n+1):
        ergebnis=ergebnis*i
    for i in range(1,n-k+1): 
        ergebnis=ergebnis//i     
    return ergebnis 
bino(49,6)
```

```python
# Das Folgende definiert eine einfache Turtle und öffnet die Zeichenfläche
from ipycanvas import Canvas
import math
class CanvasTurtle:
    def __init__(self, canvas, x=None, y=None, heading=0, color="black"):
        self.c = canvas
        self.width = canvas.width
        self.height = canvas.height
        self.x = x if x is not None else self.width / 2
        self.y = y if y is not None else self.height / 2
        self.heading = heading  # degrees, 0 = east
        self.pen = True
        self.color(color)
    # --- drawing control ---
    def penup(self):
        self.pen = False
    def pendown(self):
        self.pen = True
    def color(self, color):
        self.c.stroke_style = color
    def clear(self):
        self.c.clear()
        self.x = self.width/2
        self.y = self.height/2
        self.heading = 0 
        self.pen = True
        self.color("black")
    def _to_canvas(self, x, y):
        return self.width/2 + x, self.height/2 - y
    def _to_turtle(self, x, y):
        return x - self.width/2, self.height/2 - y
    # --- movement ---
    def forward(self, d):
        x2 = self.x + d * math.cos(math.radians(self.heading))
        y2 = self.y + d * math.sin(math.radians(self.heading))
        if self.pen:
            self.c.stroke_line(self.x, self.y, x2, y2)
        self.x, self.y = x2, y2
    def backward(self, d):
        self.forward(-d)
    def goto(self, x, y):
        cx, cy = self._to_canvas(x, y)
        if self.pen:
            self.c.stroke_line(self.x, self.y, cx, cy)
        self.x, self.y = cx, cy
    def getX(self): return self.x - self.width/2
    def getY(self): return self.height/2 - self.y
    # --- turning ---
    def left(self, a):
        self.heading -= a
    def right(self, a):
        self.heading += a
    def setheading(self, a):
        self.heading = a
canvas = Canvas(width=400, height=400)
t = CanvasTurtle(canvas)
```

```python
display(canvas)
# Ein Quadrat von Hand
t.clear()
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
t.forward(100)
t.left(90)
```

```python
# Besser mit einer Funktionsdefinition
def Quadrat(a):
    for i in range(4):
        t.forward(a)
        t.left(90)

t.clear()
t.goto(100,-100)
t.setheading(45)
t.color('blue')
Quadrat(50)
t.color("green")
t.penup()
t.goto(-50,50)
t.pendown()
t.color('red')
t.setheading(0)
Quadrat(100)
```

```python
display(canvas)
def koch(laenge,stufe): # stufe<=1
    if stufe==0:
        t.forward(laenge)
    else:
        koch(laenge/3,stufe-1) 
        t.left(60) 
        koch(laenge/3,stufe-1) 
        t.right(120) 
        koch(laenge/3,stufe-1) 
        t.left(60) 
        koch(laenge/3,stufe-1)
t.clear(); t.penup(); t.goto(-50,0); t.pendown()
koch(200,4)
```

```python
def qs(n,m):
    t.forward(n) 
    t.left(90) 
    if n<m: qs(n+10,m)
t.clear(); t.goto(-50,0)
qs(10,200)
```

```python
def baum(l,n):
    if n==0:
        t.forward(l); t.backward(l) 
        return 
    else:
        t.forward(l); t.left(25)
        baum(l-3,n-1); t.right(50)
        baum(l-3,n-1); t.left(25) 
        t.backward(l)
t.clear(); t.penup(); t.goto(-50,0); t.pendown()
baum(40,5)
```

```python
def dreieck(a):
    for i in range(3):
        t.forward(a)
        t.left(120)
def DF(L,n):
    if n==0 or L<10: 
        dreieck(L)
    else:
        dreieck(L); t.forward(5)
        t.left(60); t.forward(5); t.right(60) 
        DF(L-15,n-1)
t.clear(); t.penup(); t.goto(-50,0); t.pendown()
DF(200,10)
```
